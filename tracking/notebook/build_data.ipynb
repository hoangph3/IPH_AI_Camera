# Augmentation

```python
import cv2
import numpy as np
import os
```

```python
def augment(img_path, save_path):
    # Load image
    img = cv2.imread(img_path)
    filename = os.path.basename(img_path)

    # Apply augmentation techniques to create new images
    flipped_img = cv2.flip(img, 1)

    # Save new images
    new_filename_1 = "flipped_" + filename
    new_img_path_1 = os.path.join(save_path, new_filename_1)

    cv2.imwrite(new_img_path_1, flipped_img)
```

# WOB 

```python
import os
import re
from pathlib import Path


data_dir = "/media/hoang/Data/aihub/tracking/raw/WoB-ReID/WB_WoB-ReID/"
saved_data_dir = "/media/hoang/Data/aihub/tracking/WoB"


for ds in os.listdir(data_dir):
    ds = os.path.join(data_dir, ds)
    for scenario in os.listdir(ds):
        scenario = os.path.join(ds, scenario)

        for path in sorted(Path(scenario).glob('*.jpg')):
            name = path.resolve().name
            user_id = str(int(name.split('_')[0]))
            user_dir = Path(saved_data_dir).resolve() / user_id

            os.makedirs(user_dir, exist_ok=True)
            os.system('cp "{}" {}'.format(str(path), user_dir))
```
